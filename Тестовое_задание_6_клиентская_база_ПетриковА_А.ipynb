В стобцах dflt_price, cost_price, prdct_cnt присутствуют запятые, которые не позволят привести данные к формату соответствующему данным в столбце. 

Напишем функцию для замены запятых на точки. 

```
def normal_string(string):
  dict_month = {'янв.':'01',
                'февр.':'02',
                'мар.':'03',
                'апр.':'04',
                'мая':'05',
                'июн.':'06',
                'июл.':'07',
                'авг.':'08',
                'сент.':'09',
                'окт.':'10',
                'нояб.':'11',
                'дек.':'12'}

  try: return float(re.search('[0-9]+',string).group() + "." + dict_month[re.search('[а-я.]+',string).group()])
  except: return float(string)
```

```
data['ttl_order_cost'] = data['ttl_order_cost'].apply(normal_string)
data['dflt_price'] = data['dflt_price'].apply(normal_string)
data['cost_price'] = data['cost_price'].apply(normal_string)
data['prdct_id'] = data['prdct_id'].astype('int64')
data['prdct_cnt'] = data['prdct_cnt'].astype('int64')
data['card_nmb'] = data['card_nmb'].astype('str')
data['trnsct_date'] = data['trnsct_date'].astype('datetime64[D]')
data.info()
```

```
data['month_trnsct'] = data['trnsct_date'].astype('datetime64[M]')
```

```
montly_user_data = data.groupby(['card_nmb','month_trnsct']).agg(
                                                                 Количество_посещений_в_месяц = ('trnsct_date','nunique'),
                                                                 Выручка_от_покупателя = ('cost_price', 'sum'),
                                                                 Выручка_от_покупателя_без_учёта_скидки = ('dflt_price', 'sum'),
                                                                 Количество_чеков = ('check_nmb', 'nunique'),
                                                                 Начисленные_бонусы = ('cost_price','sum')         
                                                                 )
montly_user_data["Начисленные_бонусы"] = montly_user_data["Начисленные_бонусы"]*0.01
montly_user_data = montly_user_data.reset_index()
montly_user_data
```

```
def castumer_group(string):
    if string < 2000:
      return "Группа 1"
    elif string >= 2000 and string < 4000:
      return "Группа 2"
    elif string >= 4000 and string < 8000:
      return "Группа 3"
    else:
      return "Группа 4"
```
